–ü—Ä–æ—Å—Ç–µ–π—à–∞—è –º–æ–¥–µ–ª—å ***–ú–∞–ª—å—Ç—É—Å–∞*** —Ä–æ—Å—Ç–∞ –Ω–∞—Å–µ–ª–µ–Ω–∏—è –ó–µ–º–ª–∏:
$$ ùë¶‚Ä≤=‚àíùëòùë¶ $$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(y,t):
    k=0.6
    return -k*y
```

```python
h=1/100
t = np.arange(0,10+h,h)

solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
```

### –ú–µ—Ç–æ–¥ –≠–π–ª–µ—Ä–∞

```python
def Euler(f, y0, t):
    y=[0]*len(t)
    y[0]=1
    
    for i in range(1,len(t)):
        y[i]=y[i-1]+h*f(y[i-1],t[i-1])
    return y 
```

```python
%matplotlib inline

plt.plot(t, Euler(dydt,1,t))
plt.plot(t, solve, label = r"$y' = -ky$", color="green")
plt.grid(True)
plt.legend()
plt.show()
```
