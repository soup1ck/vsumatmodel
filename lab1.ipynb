# Номер 1 
## Граков В А КФА
### x' = kx 
### Жесткая модель Мальтуса. Скорость роста зависит от коэф k и чем он больше, тем быстрее растет население, что быстрее приведет к катастрофе.

```python
#Номер 1
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
    k=0.42
    return k*x

t=np.linspace(0, 10, 100)
solve = odeint(dydt, 1 , t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t, solve, label=r"x'=kx")
plt.grid(True)
plt.legend()
plt.show()
```

# Номер 2
## Граков В А КФА
### x' = k(x)x 
### Мягкая модель Мальтуса. Если мы добавим коэф размножения k(x), то модель становится мягкой, что позволяет нам более точно оценить возможность катастрофы. 

```python
#Номер 2
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def k(x):
    a=4
    b=18
    return a-b*x

def dydt(x,t):
    return k(x)*x

t=np.linspace(0, 10, 100)
solve = odeint(dydt, 1 , t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t, solve, label=r"x'=k(x)x")
plt.grid(True)
plt.legend()
plt.show()
```

# Номер 3
## Граков В А КФА
### x' = x - x^2 - c
### Жесткая модель рыболовства. Чем больше интенсивность рыболовста c,  тем меньше популяция рыб, для нашей модели критическое значение с = 1/4. Заметим, что есть такое состояние, когда мы можем при определенных парметрах c добиться того, что со временем популяция рыб будет полностью уничтожена.

```python
#Номер 3
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
    c=1
    return x-x*x-c

t=np.linspace(0, 10, 100)
solve = odeint(dydt, 1 , t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t, solve, label=r"x'=x - x^2 - c")
plt.grid(True)
plt.legend()
plt.show()
```

# Номер 4
## Граков В А КФА
### x' = x - g(x)x- c
### Это мягкая модель предыдущего номера, здесь критическое значение достигается при максимуме функции g(x)x.

```python
#Номер 4
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def g(x):
    a=1
    b=2
    return a*x - b

def dydt(x,t):
    c=1
    return x-g(x)*x -c 

t=np.linspace(0, 10, 100)
solve = odeint(dydt, 1 , t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t, solve, label=r"x'=x - g(x)x - c")
plt.grid(True)
plt.legend()
plt.show()
```

# Номер 5
## Граков В А КФА
### x' = x - x^2 - kx
### Мягкая модель с дифф квотой. k - дифференциальня квота, которая подлежит выбору. При k<1 с течением времени устанавливается стационарное состояние, которое устойчиво.

```python
#Номер 5
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
    k=1
    return x - x*x - k*x 

t=np.linspace(0, 10, 100)
solve = odeint(dydt, 1 , t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t, solve, label=r"x'=x - x^2 - kx")
plt.grid(True)
plt.legend()
plt.show()
```
