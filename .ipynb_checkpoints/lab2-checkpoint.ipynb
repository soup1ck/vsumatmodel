```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
from scipy.integrate import odeint 
data = {1986:1210000,
        1987:1231000,
        1988:1251000,
        1989:1264621,
        1990:1204000,
        1991:1275000,
        1992:1271000,
        1993:1258000,
        1994:1247000,
        1995:1278000,
        1996:1276000,
        1997:1175000,
        1998:1272000,
        1999:1272900,
        2000:1266300,
        2001:1256900,
        2002:1293537,
        2003:1293500,
        2004:1334400,
        2005:1304300,
        2006:1308400,
        2007:1375100,
        2008:1383000,
        2009:1332264,
        2010:1399772,
        2011:1350100,
        2012:1377738,
        2013:1398074,
        2014:1412346,
        2015:1428042,
        2016:1444439,
        2017:1455904,
        2018:1468833,
        2019:1483119,
        2020:1493749
}
years=data.keys()
people=data.values()
years=np.array(list(years))
people=np.array(list(people))

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
    k=0.02
    return k * x

solve = odeint(dydt, y0=1000000, t=years)
solve = solve.flatten()

plt.plot(years, solve, label = r"Мягкая модель Мальтуса")
plt.plot(years, people, color="red", label = r"Население Казани")
plt.grid(True)
plt.legend()
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def k(x):
    a=400000000
    b=350
    return a - b*x

def dydt(x,t):
    return k(x)*x

solve = odeint(dydt, y0=100000, t=years)
solve = solve.flatten()

plt.plot(years, solve, label = r"Жёсткая модель", color = "green")
plt.plot(years, people, color="red", label = r"Население Казани")
plt.grid(True)
plt.legend()
plt.show()
```
